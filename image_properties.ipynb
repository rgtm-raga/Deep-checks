## Calcuation of Image proprties

['Area', 'Aspect Ratio', 'Brightness', 'Mean Blue Relative Intensity', 'Mean Green Relative Intensity', 'Mean Red Relative Intensity', 'RMS Contrast']<br>
Image Sequence: RGB

```python
import cv2
import os
from PIL import Image
import numpy as np
import pandas as pd
from PIL.ImageStat import Stat
```

```python
def area(image):
    image=np.array(image)
    height,width,_=image.shape
    return height*width
    
def aspect_ratio(image):
    image=np.array(image)
    height,width,_=image.shape
    return height/width

def brightness(image):
    image=np.uint8(image*255)
    image=Image.fromarray(image)
    image=image.convert('L')
    return Stat(image).mean[0]

def mbr_intesity(image):
    # image=np.array(image)/255
    height,width,_=image.shape
    imb=np.zeros(image.shape) #
    for i in range(height):
        for j in range(width):
            imb[i,j]=image[i,j,2]/sum(image[i,j,:]) # red
    return np.sum(imb)/(height*width) #
def mgr_intesity(image):
    # image=np.array(image)/255
    height,width,_=image.shape
    img=np.zeros(image.shape) #
    for i in range(height):
        for j in range(width):
            img[i,j]=image[i,j,1]/sum(image[i,j,:]) # red
    return np.sum(img)/(height*width)

def mrr_intesity(image):
    # image=np.array(image)/255
    height,width,_=image.shape
    imr=np.zeros(image.shape) #
    for i in range(height):
        for j in range(width):
            imr[i,j]=image[i,j,0]/sum(image[i,j,:]) # red
    return np.sum(imr)/(height*width)

def rms_contrast(image):
    image=np.uint8(image*255)
    img_grey=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    return img_grey.std()
```

```python
properties=['Area', 'Aspect Ratio', 'Brightness', 'Mean Blue Relative Intensity', 'Mean Green Relative Intensity', 'Mean Red Relative Intensity', 'RMS Contrast']
image_properties=[]
for img in os.listdir("red"):
    image_property=[]
    img_path=os.path.join("red",img)
    # image=Image.open(img_path)
    image=cv2.imread(img_path)/255
    image_property.append(area(image))
    image_property.append(aspect_ratio(image))
    image_property.append(brightness(image))
    image_property.append(mbr_intesity(image))
    image_property.append(mgr_intesity(image))
    image_property.append(mrr_intesity(image))
    image_property.append(rms_contrast(image))
    image_properties.append(image_property)
red_feature=np.array(image_properties)
red_df=pd.DataFrame(red_feature,columns=properties)

image_properties=[]
for img in os.listdir("green"):
    image_property=[]
    img_path=os.path.join("green",img)
    # image=Image.open(img_path)
    image=cv2.imread(img_path)/255
    image_property.append(area(image))
    image_property.append(aspect_ratio(image))
    image_property.append(brightness(image))
    image_property.append(mbr_intesity(image))
    image_property.append(mgr_intesity(image))
    image_property.append(mrr_intesity(image))
    image_property.append(rms_contrast(image))
    image_properties.append(image_property)               

green_feature=np.array(image_properties)
green_df=pd.DataFrame(green_feature,columns=properties)
```

```python
def img_prop(path):
    properties=['Area', 'Aspect Ratio', 'Brightness', 'Mean Blue Relative Intensity', 'Mean Green Relative Intensity', 'Mean Red Relative Intensity', 'RMS Contrast']
    image_properties=[]
    for img in os.listdir(path):
        if img!='.ipynb_checkpoints':
            image_property=[]
            img_path=os.path.join(path,img)
            # image=Image.open(img_path)
            image=cv2.imread(img_path)/255
            image_property.append(area(image))
            image_property.append(aspect_ratio(image))
            image_property.append(brightness(image))
            image_property.append(mbr_intesity(image))
            image_property.append(mgr_intesity(image))
            image_property.append(mrr_intesity(image))
            image_property.append(rms_contrast(image))
            image_properties.append(image_property)
    red_feature=np.array(image_properties)
    return pd.DataFrame(red_feature,columns=properties)

def join_df(*df,axis=None):
    return pd.concat([dfs for dfs in df],axis=axis)
```
